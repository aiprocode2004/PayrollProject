```python
class Employee:
    def __init__(self, name):
        self.name = name

    def calculate_net_pay(self):
        pass

    def calculate_bonus(self):
        return 0


class HourlyContractor(Employee):
    def __init__(self, name, hours, rate):
        super().__init__(name)
        self.hours = hours
        self.rate = rate

    def calculate_net_pay(self):
        return self.hours * self.rate


class SalariedEmployee(Employee):
    def __init__(self, name, salary):
        super().__init__(name)
        self.salary = salary

    def calculate_net_pay(self):
        return self.salary

    def calculate_bonus(self):
        return self.salary * 0.10


class PayrollSystem:
    def __init__(self):
        self.employees = []

    # Add Hourly Contractor
    def add_hourly(self):
        name = input("Enter Name: ")
        hours = int(input("Hours Worked: "))
        rate = float(input("Hourly Rate: "))

        self.employees.append(HourlyContractor(name, hours, rate))
        print("Hourly Contractor Added!")

    # Add Salaried Employee
    def add_salary(self):
        name = input("Enter Name: ")
        salary = float(input("Monthly Salary: "))

        self.employees.append(SalariedEmployee(name, salary))
        print("Salaried Employee Added!")

    # View Employees
    def view_employees(self):
        if not self.employees:
            print("\nNo Employees Found!")
            return

        print("\n------ EMPLOYEE LIST ------")
        for emp in self.employees:
            print(f"Type : {type(emp).__name__}")
            print(f"Name : {emp.name}")
            print("--------------------------")

    # Process Payroll
    def process_monthly_payroll(self):
        if not self.employees:
            print("\nNo Employees Found!")
            return

        print("\n------ MONTHLY PAYROLL ------")

        for emp in self.employees:
            pay = emp.calculate_net_pay()
            bonus = emp.calculate_bonus()

            print(f"\nEmployee : {emp.name}")
            print(f"Type     : {type(emp).__name__}")
            print(f"Net Pay  : {pay}")
            print(f"Bonus    : {bonus}")

    # Save Payroll Report
    def save_report(self):
        with open("payroll.txt", "a") as file:
            file.write("MONTHLY PAYROLL REPORT\n")
            file.write("-------------------------\n")

            for emp in self.employees:
                file.write(f"Type : {type(emp).__name__}\n")
                file.write(f"Name : {emp.name}\n")
                file.write(f"Net Pay : {emp.calculate_net_pay()}\n")
                file.write(f"Bonus : {emp.calculate_bonus()}\n")
                file.write("-------------------------\n")

        print("Payroll report saved in payroll.txt")


# Main Program
system = PayrollSystem()

while True:
    print("\n===== PAYROLL MANAGEMENT SYSTEM =====")
    print("1. Add Hourly Contractor")
    print("2. Add Salaried Employee")
    print("3. View Employees")
    print("4. Process Monthly Payroll")
    print("5. Save Payroll Report")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        system.add_hourly()

    elif choice == "2":
        system.add_salary()

    elif choice == "3":
        system.view_employees()

    elif choice == "4":
        system.process_monthly_payroll()

    elif choice == "5":
        system.save_report()

    elif choice == "6":
        print("Thank You!")
        break

    else:
        print("Invalid Choice!")
```
